# MatplotLib & Seaborn

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

## Line chart

```python
apples = [0.895, 0.91, 0.919, 0.925, 0.929, 0.933]
plt.plot(apples)
```

```python
# Customizing the x-axis
years = [2018, 2019, 2020, 2021, 2022, 2023]
plt.plot(years, apples)
```

```python
# Adding Labels
plt.xlabel('Year')
plt.ylabel('Yield per hectare')
plt.plot(years, apples)
```

```python
# Plotting multiple lines
year = range(2000, 2010)
apple = [0.895, 0.91, 0.919, 0.925, 0.929, 0.933, 0.939, 0.941, 0.948, 0.950]
orange = [0.965, 0.969, 0.959, 0.955, 0.949, 0.943, 0.939, 0.931, 0.928, 0.910]

plt.plot(year, apple)
plt.plot(year, orange)
```

```python
# Chart Title and Legend
plt.plot(year, apple)
plt.plot(year, orange)

plt.xlabel('Year')
plt.ylabel('Yield per hectare')

plt.title('Crop yields in Kanto')
plt.legend(['Apples', 'Oranges'])
```

```python
# Using markers
plt.plot(year, apple, marker='o')
plt.plot(year, orange, marker='x')

plt.xlabel('Year')
plt.ylabel('Yield per hectare')

plt.title('Crop yields in Kanto')
plt.legend(['Apples', 'Oranges'])
```

```python
# Styling Markers
plt.plot(year, apple, marker='s', c='b', ls='-', lw=2, ms=8, mew=2, mec='navy')
plt.plot(year, orange, marker='o', c='r', ls='--', lw=3, ms=10, alpha=.5)

plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')

plt.title("Crop Yields in Kanto")
plt.legend(['Apples', 'Oranges']);
```

```python
# Using FMT
# fmt = [marker][line][color]
plt.plot(year, apple, 's-b')
plt.plot(year, orange, 'o--r')

plt.xlabel('Year')
plt.ylabel('Yield per hectare')

plt.title('Crop yields in Kanto')
plt.legend(['Apples', 'Oranges'])
```

```python
# Change size
plt.figure(figsize=(12, 6))
plt.plot(year, orange)
```

```python
plt.figure(figsize=(3, 3))
plt.plot(year, orange)
```

```python
# Improving using seaborn
sns.set_style('whitegrid')
plt.plot(year, apple, 's-b')
plt.plot(year, orange, 'o--r')

plt.xlabel('Year')
plt.ylabel('Yield per hectare')

plt.title('Crop yields in Kanto')
plt.legend(['Apples', 'Oranges'])
```

```python
sns.set_style('darkgrid')
plt.plot(year, apple, 's-b')
plt.plot(year, orange, 'o--r')

plt.xlabel('Year')
plt.ylabel('Yield per hectare')

plt.title('Crop yields in Kanto')
plt.legend(['Apples', 'Oranges'])
```

```python
# Set default styles
import matplotlib
matplotlib.rcParams['font_size'] = 12
matplotlib.rcParams['figure.figsize'] = (9,5)
```

## Scatter Plot

```python
flowers = sns.load_dataset('iris')
flowers
```

```python
flowers.species.unique()
```

```python
plt.plot(flowers.sepal_length, flowers.sepal_width)
```

```python
sns.scatterplot(x=flowers.sepal_length, y=flowers.sepal_width)
```

```python
sns.scatterplot(x=flowers.sepal_length, y=flowers.sepal_width, hue=flowers.species, s=100)
```

```python
plt.figure(figsize=(12, 6))
plt.title('Sepal Dimensions')

sns.scatterplot(x=flowers.sepal_length, y=flowers.sepal_width, hue=flowers.species, s=100)
```

## Histogram

```python
flowers.sepal_width
```

```python
flowers.describe()
```

```python
plt.title('Distribution of Sepal Width')
plt.hist(flowers.sepal_width)
```

```python
# Controlling the number and size of bins
plt.title('Distribution of Sepal Width')
plt.hist(flowers.sepal_width, bins=np.arange(2, 5, 0.25))
```

```python
# Multiple Histograms
setosa = flowers[flowers.species == 'setosa']
versicolor = flowers[flowers.species == 'versicolor']
virginica = flowers[flowers.species == 'virginica']
```

```python
plt.hist(setosa.sepal_width, alpha=0.4, bins=np.arange(2, 5, 0.25))
plt.hist(virginica.sepal_width, alpha=0.4, bins=np.arange(2, 5, 0.25))
plt.legend(['Setosa', 'Virginica'])
```

```python
# Stacking Histograms Together
plt.title('Distribution of Sepal Width')
plt.hist([setosa.sepal_width, virginica.sepal_width, versicolor.sepal_width],
         bins=np.arange(2, 5, 0.25), stacked=True)
plt.legend(['Setosa', 'Virginica', 'Versicolor'])
```

## Bar Chart

```python
year = range(2000, 2010)
apple = [0.895, 0.91, 0.919, 0.625, 0.929, 0.733, 0.939, 0.841, 0.948, 0.750]
orange = [0.965, 0.969, 0.959, 0.955, 0.949, 0.943, 0.939, 0.931, 0.928, 0.910]
plt.plot(year, apple)
```

```python
plt.bar(year, apple)
plt.plot(year, apple, 'o--r')
plt.title('Yield of Apples')
```

```python
plt.bar(year, apple)
plt.bar(year, orange, bottom=apple)
```

```python
# Bar plot with averages
tips = sns.load_dataset('tips')
tips
```

```python
avg_bill = tips.groupby('day')[['total_bill']].mean()
avg_bill
```

```python
plt.bar(avg_bill.index, avg_bill.total_bill)
```

```python
sns.barplot(x='day', y='total_bill', data=tips)
```

```python
sns.barplot(x='day', y='total_bill', hue='sex', data=tips)
```

```python
sns.barplot(x='day', y='total_bill', hue='smoker', data=tips)
```

```python
# Horizontal Bars
sns.barplot(x='total_bill', y='day', hue='sex', data=tips)
```

## Contour Plot

```python
plt.title("Flowers")

sns.kdeplot(x=flowers.sepal_length, y=flowers.sepal_width, fill=True, thresh=False);
```

```python
setosa = flowers[flowers.species == 'setosa']
virginica = flowers[flowers.species == 'virginica']

plt.title("Flowers (Setosa & Virginica)")

sns.kdeplot(x=setosa.sepal_length, y=setosa.sepal_width, fill=True, cmap='Reds', thresh=0.1)
sns.kdeplot(x=virginica.sepal_length, y=virginica.sepal_width, fill=True, cmap='Blues', thresh=0.1);
```

## Box Plot

```python
# Chart title
plt.title("Daily Total Bill")

# Draw a nested boxplot to show bills by day and time
sns.boxplot(x=tips.day, y=tips.total_bill, hue=tips.smoker);
```

## Heatmap

```python
flights = sns.load_dataset('flights')
flights
```

```python
plt.plot(flights.passengers)
```

```python
flights = flights.pivot('month', 'year', 'passengers')
flights
```

```python
plt.title('Heatmap of Passengers')
sns.heatmap(flights)
```

```python
plt.title('Heatmap of Passengers')
sns.heatmap(flights, fmt='d', annot=True, cmap='Blues')
```

## Plotting multiple charts in a grid


```python
plt.subplots(2, 3)
```

```python
plt.subplots(2, 3)
plt.tight_layout(pad=2)
```

```python
figs, axes = plt.subplots(2, 3, figsize=(12, 9))

axes[0, 0].plot(year, orange, 'o--r')
axes[0, 0].plot(year, apple, 's-b')
axes[0, 0].set_title('Yield of Oranges')
axes[0, 0].set_xlabel('Year')
axes[0, 0].set_ylabel('Oranges')

axes[0, 1].set_title('Sepal Length vs Width')
sns.scatterplot(x=flowers.sepal_length, y=flowers.sepal_width, hue=flowers.species, s=100, ax=axes[0, 1])


axes[0, 2].set_title('Distribution of Sepal Width')
axes[0, 2].hist([setosa.sepal_width, virginica.sepal_width, versicolor.sepal_width],
         bins=np.arange(2, 5, 0.25), stacked=True)
axes[0, 2].legend(['Setosa', 'Virginica', 'Versicolor'])

axes[1, 0].set_title('Yield (apple vs oranges)')
axes[1, 0].bar(year, apple)
axes[1, 0].bar(year, orange, bottom=apple)

axes[1, 1].set_title('Passenger Data')
sns.heatmap(flights, ax=axes[1, 1])

axes[1, 2].imshow(img)
```
